```python
from PIL import Image, ImageDraw
import torchvision.transforms as T
from lxml import etree
import os


xml = etree.parse('')

dict_of_polygons = {}
for img_tag in xml.findall('image'):
    
    result_polygons = []
    
    image_name = img_tag.get('name')[:-4]
   
    for polygon in img_tag.findall('polygon'):
       
        dots = []
        for dot in polygon.get('points').split(';'):
            dot = (float(dot.split(',')[0]), float(dot.split(',')[1]))
            dots.append(dot)
       
        result_polygons.append(dots)
        dict_of_polygons[image_name] = result_polygons
```

```python
# image = Image.open('/Users/aleksandrrubashevskii/Desktop/cvat_work/P_o001_L_S1_Nr1.bmp')
reference_to_folder = '/Users/aleksandrrubashevskii/Desktop/dataset_90/' 
img_names = os.listdir(reference_to_folder + 'img_5/')

for img_name in img_names:
    
    if img_name != '.DS_Store':

        image = Image.open(reference_to_folder + 'img_5/' + img_name)

        true_mask = Image.new('L', (image.size[0], image.size[1]), 0)
        
        for i in range(len(dict_of_polygons[img_name[:-4]])):
            draw = ImageDraw.Draw(true_mask).polygon(dict_of_polygons[img_name[:-4]][i], outline=255, fill=255)
        
        true_mask.save(reference_to_folder + 'mask/' + 'mask_' + img_name[6:][:-4] + '.png')
```
